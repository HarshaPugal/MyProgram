```python
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
import pandas as pd
x=[7,17,27,37]
y=[6,16,26,36]
f=plt.figure(figsize=(7,5),facecolor='wheat',edgecolor='black',linewidth=5)
plt.plot(x,y)
plt.title("Linear graph",fontsize=20,color="maroon")
plt.ylabel('Y-Axis',color="black",fontsize=15)
plt.xlabel('X-Axis',color="black",fontsize=15)
plt.ylim(0,50)
plt.xlim(0,50)
plt.legend(['Sample'])
plt.show()
```

```python
data={'CSE':31,'ECE':20,'CIVIL':20,'EEE':10}
department=list(data.keys())
students=list(data.values())
f=plt.figure(figsize=(7,5),facecolor="wheat")
plt.bar(department,students,color="maroon",width=0.4)
plt.show()
```
