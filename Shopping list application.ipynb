```python
"""
Excersise number:09
Start date      :18/08/2023
End date        :18/08/2023
Description     :To create shopping list using list datastructure
"""
shopping_list=[]
while(True):
    print("\n\n1.Press 1 to add an item to your shopping list\n2.Press 2 to remove an item from your shopping list")
    print("3.Press 3 to replace an item in your shopping list\n4.Press 4 to View your shopping list")
    print("5.Press 5 to exit")
    Choice=int(input(""))
    if(Choice==1):  #Add an item to the list
        Add_item=input("Enter an item to add it to your shopping list")
        shopping_list.append(Add_item)
    elif(Choice==2):  #Remove an item from the list
        rind=0
        Remove_item=input("Enter an item to remove it from your shopping list")
        for i in range(len(shopping_list)):   #To find the index of the item to be removed
            if(shopping_list[i]==Remove_item):
                rind=i
            else:
                i=i+1
        del shopping_list[rind]
    elif(Choice==3):    #Replace an item in the list
        ind=0
        Old_item=input("Enter the old item to replace")
        New_item=input("Enter the new item")
        for i in range(len(shopping_list)):   #To find the index of the old item to be replaced with new item
            if(shopping_list[i]==Old_item):
                ind=i
            else:
                i=i+1
        shopping_list[ind]=New_item
    elif(Choice==4):   #View the item in the list
        print(shopping_list[:])
    elif(Choice==5):  
        break
```
